1.2 Напишите функцию `show_info`, которая для рецепта по его `id` создает строку (в смысле объекта python) с описанием следующего вида:

```
"Название"

1. Шаг 1.
2. Шаг 2.
----------
#тэг1 #тэг2
```

    
Данные для создания строки получите из файлов `recipes_sample_with_tags_ingredients.csv`, `steps_sample.xml` (__ЛР4__) и `tags_sample.csv` (__ЛР5__). 
Выведите созданную строку на экран.

````python
with open('steps_sample.xml') as file:
    xml_data = BeautifulSoup(file, 'xml')
    
with open('tags_sample.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    head = next(reader)
    tags_sample = {}
    for i in reader:
        if i[0] not in tags_sample.keys():
            tags_sample[i[0]] = []
        tags_sample[i[0]].append(i[1])
        
recipes_ingredients = pd.read_csv('recipes_sample_with_tags_ingredients.csv')
````

````python
def show_info(rec_id):
    rec_name = (re.sub(r'\s+', ' ', recipes_ingredients[recipes_ingredients['id'] == rec_id].values[0][1])).capitalize()
    for recipe in xml_data.find_all('recipe'):
        rec_found = recipe.find('id')
        if int(rec_found.text) == rec_id:
            steps = recipe.find_all("step")
            
    result_str = '"' + rec_name + '"' + "\n\n"
    for i in range(len(steps)):
        new_str = steps[i].text.capitalize()
        result_str += f'{i+1}. {new_str}.\n'
        
    result_str += f'{"-"*70}\n'
    
    for k, v in tags_sample.items():
        if int(k) == rec_id:
            tags = []
            for i in v:
                tags.append(f'#{i} ')
                
    for tag in tags:
        result_str += tag
    
    return result_str
````

2.1 Посчитайте кол-во отзывов, в которых встречаются числа.

````python
pattern = re.compile(r'.*\d.*')
q = 0
for i in reviews_in_list:
    res = re.findall(pattern, i)
    if len(res) != 0:
        q += 1
print(q)
````

2.2 Найдите все смайлики в отзывах. Смайлик состоит из трех частей: глаза (символ `=` или `:`), нос (символ `-`), губы (символ `)` или `(`). Смайлик может иметь вид "глаза-нос-губы" или "губы-нос-глаза". Нос может отсутствовать.

````python
pattern = re.compile(r'([=:]-?[()])|([()]-?[=:])')
q = 0
for i in reviews_in_list:
    res = re.findall(pattern, i)
    if len(res) > 0:
        print(i)
        print(f'\n конструкция - {res[0][0]}')
        print(f'{"-"*125}\n')
        q += 1
print(q)
````

````python
stroka_dict = []
otzivi = []
for i in recipes['description']:
  for j in nltk.sent_tokenize(str(i)):
    words = nltk.word_tokenize(j)
    words=[word for word in words if word.isalpha()]#убирает знаки препинания   Метод isalpha() возвращает True, если все символы в строке являются алфавитными. Если нет, возвращается False.
    stroka_dict.append(words)
  otzivi.append(stroka_dict)
  stroka_dict =[]
````

3.2 Посчитайте кол-во уникальных слов среди всех отзывов (без учета регистра и знаков препинания).

````python
#Этот токенизатор разбивает исходный текст на подстроки, используя регулярное выражение, передаваемое ему в качестве параметра. 
#первый параметр регулярное выражение, в нашем случае идет на вход поиск всех слов, без знаков препинания
pattern = '[a-r]\w+'#поиск всех слов начинающихся с маленькой буквы, в нашем случае не всех слов, как мы без учета регистра и приводим в нижний
tokenizer = RegexpTokenizer(pattern)
````

````python
All_string = []
for i in recipes["description"].fillna(""):
  All_string.append(i)
All_atring = ''.join(All_string)
fdist = FreqDist(tokenizer.tokenize(All_atring.lower()))# подаем в tokenizer.tokenize одну большую строчку(tokenizer.tokenize не берет в себя знаки препинания), а потом FreqDist уже посчитает колличество уникальных слов
````

````python
print("Колличество всех слов: ", fdist.N())#Return the total number of sample values (or “bins”) that have counts greater than zero.
print("Колличество униккальных слов: ", fdist.B())#Return the total number of sample outcomes that have been recorded by this FreqDist.
````

````python
fdist.most_common(10)
````
